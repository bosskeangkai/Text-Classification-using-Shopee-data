#### Scraping Product name from Shopee

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import bs4
import time
import re
import pandas as pd
import numpy as np

pd.options.display.max_colwidth=300
```

```python
options = Options()
driver = webdriver.Chrome(executable_path=r"C:\Users\Keangkai\Desktop\dev\side_project\Shopee_web_scraping\chromedriver_win32\chromedriver")
driver.get('https://shopee.co.th/')
time.sleep(5)

thai_button=driver.find_element('xpath','/html/body/div[2]/div[1]/div[1]/div/div[3]/div[1]/button')
thai_button.click()
time.sleep(3)

close_adver = driver.execute_script('return document.querySelector("shopee-banner-popup-stateful").shadowRoot.querySelector("div.shopee-popup__close-btn")')
close_adver.click()
time.sleep(3)

search = driver.find_element('xpath','//*[@id="main"]/div/header/div[2]/div/div[1]/div[1]/div/form/input')
search.send_keys('เสื้อผ้าผู้ชาย')
#กด Enter
search.send_keys(Keys.ENTER)

#Zoom out เพื่อให้ load ทั่วหน้าจอ เนื่องจากหน้าเว็บของ Shopee จะไม่ load ทั้งหมดจนกว่าจะ scroll down ลงไป
driver.execute_script("document.body.style.zoom='10%'")
time.sleep(5)

data = driver.page_source #ดึงข้อมูลจากหน้าเว็บ
soup = bs4.BeautifulSoup(data) #จัดในรูปแบบ BeautifulSoup
```

```python
elements=soup.select(".row>div")
elements
```

```python
records=[]
for i in elements:
    product_name_discount=soup.find_all('div',{'class':'dpiR4u'})
    for product_dis in product_name_discount:
        records.append(product_dis.text)
df=pd.DataFrame(records,columns=['product'])
df
```
