```python
raw_data=pd.read_excel('describe_name.xlsx')
raw_data.head(10)
```

```python
stopwords_removed=[]
for product in raw_data['describe_product']:
    word_list=deepcut.tokenize(product)
    name_with_no_stopwords=""
    for word in word_list:
        if word in stopwords:
            continue
        name_with_no_stopwords+=""
        name_with_no_stopwords += word
    stopwords_removed.append(name_with_no_stopwords)
```
