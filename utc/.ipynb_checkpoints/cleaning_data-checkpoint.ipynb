##### ใช้ RegEx ในการคัดเอาตัวเลขและข้อความคำว่า "พัน" ออกมา
- สร้างตัวแปรดึง column sold_out มาจัดการ
- ใช้ split() เพื่อแยก string ออกมา ซึ่งจะ return ออกมาเป็น list
- จากนั้น indexing เข้าไปตำแหน่งที่ 1 เพื่อคัดเอาตัวเลข pattern num,num+'พัน'
- สร้าง condition เพื่อเช็ค pattern ของ str\
    ถ้าเจอคำว่า 'พัน' ใน str ให้เปลี่ยนเป็น '000' แล้วเปลี่ยน type เป็น float *1000\
    ถ้าไม่มีคำว่า 'พัน' ใน str ให้เปลี่ยนเป็น type float ได้เลย



```python
select_num=raw_data['sold_out']
select_num.head(5)
```

```python
sold_out_amount=[]
for num in select_num:
    find_thousand=num.split()[1]
    if 'พัน' in find_thousand:
        # find_thousand=int(float(find_thousand.replace('พัน','000'))*1000)
        thousand=find_thousand.replace('พัน','000')
        if '.' in thousand:
            thousand=int(float(thousand.replace('พัน','000'))*1000)
            sold_out_amount.append(thousand)
        else:
            thousand=int(thousand.replace('พัน','000'))
            sold_out_amount.append(thousand)
    else:
        find_thousand=int(find_thousand)
        sold_out_amount.append(find_thousand)
sold_out_amount
```

```python
len(sold_out_amount)
```

```python
# insert(index, column_name, data)
raw_data.insert(2,'sold_out_amount',sold_out_amount)
```

##### Split ราคาออกมาโดยทำการแยกออกมาเป็น column ใหม่
- แยกประเภท pattern ของราคา มีทั้งหมด 3 ประเภท โดยการใช้ condition
    - ฿129 - ฿149
    - ฿129
    - ฿149฿129 (*หมายถึง ลดราคาจาก 149 เหลือ 129)\
- สร้างเงื่อนไขให้แต่ละ pattern ราคา\
    ถ้าหากราคาตั้งแต่หลักพันจะมีเครื่องหมาย ','
    - ฿129 - ฿149
        - ทำการ split(' - ') --> จะได้เป็น list จากนั้น replace('฿','') append เข้าไป list ราคา 1 
        - ทำการ split(' - ') --> จะได้เป็น list จากนั้น replace('฿','') append เข้าไป list ราคา 2
     - ฿129
         - replace('฿','') append เข้าไป list ราคา 1 
         - สร้างตัวแปรใหม่ที่เก็บตัวแปรเดิม เข้าไป list ราคา 2
     - ฿149฿129 
         - ทำการ split ออกมาอาจจะใช้วิธี RegEx จากนั้น replace('฿','') append เข้าไป list ราคา 1
         - ทำการ split ออกมาอาจจะใช้วิธี RegEx จากนั้น replace('฿','') append เข้าไป list ราคา 2

```python
several_price=raw_data['price']
several_price.head(2)
```

```python
price1_list=[]
price2_list=[]

for pattern in several_price: 
    # pattern 1 :฿129 - ฿149
    if '-' in pattern:
        splited_price=pattern.split(' - ')
        for i in splited_price:
            if ',' in i:
                remove_comma=i.replace(',','')
                if splited_price.index(i)%2==0:
                    pattern1_1st=(i.replace('฿',''))
                    price1_list.append(pattern1_1st)
                else:
                    pattern1_2nd=(i.replace('฿',''))
                    price2_list.append(pattern1_2nd)
            else:
                if splited_price.index(i)%2==0:
                    pattern1_1st=(i.replace('฿',''))
                    price1_list.append(pattern1_1st)
                else:
                    pattern1_2nd=(i.replace('฿',''))
                    price2_list.append(pattern1_2nd)

    # pattern 2 :฿149฿129
    elif pattern.count('฿')==2 and '-' not in pattern:# return list ['฿149฿129']
        # pattern=re.findall(r'.*(^฿\d[0-9]{0,3}|.*(^฿\d[0-9]{0,4}\,\d[0-9]{0,4}))',pattern) # return tuple from regEx
        pattern3_split=re.sub(r'^.','',pattern).split('฿')
        for i in pattern3_split:
            if ',' in i:
                remove_comma=i.replace(',','')
                if  pattern3_split.index(i)%2==0:
                    pattern3_1st=(remove_comma)
                    price1_list.append(pattern3_1st)
                else:
                    pattern3_2nd=(remove_comma)
                    price2_list.append(pattern3_2nd)
            else:
                if  pattern3_split.index(i)%2==0:
                    pattern3_1st=(i)
                    price1_list.append(pattern3_1st)
                else:
                    pattern3_2nd=(i)
                    price2_list.append(pattern3_2nd)

    # pattern 3 :฿129
    else:
        # pattern element from several_price
        pattern2=pattern.replace('฿','')
        if ',' in pattern2:
            remove_comma=int(pattern2.replace(',',''))
            duplicate_patt2=remove_comma
            price1_list.append(remove_comma)
            price2_list.append(duplicate_patt2)
        else:
            duplicate_patt2=pattern2
            price1_list.append(pattern2)
            price2_list.append(duplicate_patt2)
```

```python
raw_data.insert(2,'first_price',price1_list)
raw_data.insert(3,'second_price',price2_list)
raw_data.head(5)
```
